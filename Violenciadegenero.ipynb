# **Trabajando con Graficos**

```python
import pandas as pd

# Cargar la base de datos
dataframe = pd.read_csv("/content/drive/MyDrive/IMD/atencion_violencia_genero.csv",index_col=0, encoding ='latin-1', sep = ";")

# Agrupar por municipio y contar el número de subcategorías únicas
conteo_subcategorias = dataframe.groupby('municipio')['subcategoria'].nunique()

# Mostrar el resultado
print(conteo_subcategorias)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
conteo_subcategorias.plot(kind='bar')
plt.xlabel('Municipio', fontsize=10)
plt.ylabel('Cantidad de Subcategorías')
plt.title('Cantidad de Subcategorías por Municipio')

plt.xticks(rotation=45, ha='right', fontsize=8)
plt.subplots_adjust(bottom=0.2)

ax = plt.gca()
xticks = ax.get_xticks()
ax.set_xticks(xticks[::2])  # Ajusta el espaciado entre las categorías del eje x
ax.set_xticklabels(conteo_subcategorias.index[::2], fontsize=6, ha='right', rotation=45)  # Ajusta el tamaño de letra y la alineación en el eje x

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar la base de datos
dataframe = pd.read_csv("/content/drive/MyDrive/IMD/atencion_violencia_genero.csv",index_col=0, encoding ='latin-1', sep = ";")

# Agrupar por municipio y contar el número de categorías únicas
conteo_categorias = dataframe.groupby('municipio')['categoria'].nunique()

# Crear el gráfico de barras en color verde
plt.figure(figsize=(12, 6))
conteo_categorias.plot(kind='bar', color='green')
plt.xlabel('Municipio')
plt.ylabel('Cantidad de Categorías')
plt.title('Cantidad de Categorías por Municipio')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar la base de datos
dataframe = pd.read_csv("/content/drive/MyDrive/IMD/atencion_violencia_genero.csv",index_col=0, encoding ='latin-1', sep = ";")

# Contar el número de categorías únicas
conteo_categorias = dataframe['categoria'].value_counts()

# Crear el gráfico de torta
plt.figure(figsize=(8, 8))
conteo_categorias.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.ylabel('')
plt.title('Distribución de Categorías')
plt.legend(labels=conteo_categorias.index, loc='best')
plt.show()
```
